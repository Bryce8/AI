```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df=pd.read_json('games.json')
plt.figure(figsize=(10, 6))
plt.scatter(df['discount'], df['positive_ratio'], color='blue', edgecolor='black')

# Adding labels and title
plt.title('Impact of Game Discount on Positive Review Percentage', fontsize=14)
plt.xlabel('Discount Amount (in %)', fontsize=12)
plt.ylabel('Positive Review Percentage', fontsize=12)

# Displaying the scatter plot
plt.grid(True)
plt.show()
plt.show()
```
