```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming the dataframe 'df' is already loaded
# Example: df = pd.read_json("tsla.json")

# Extract relevant columns for scatter plots
open_price = df['open']
rsi_7 = df['rsi_7']
rsi_14 = df['rsi_14']

# Calculate the correlation coefficients
corr_open_rsi_7 = open_price.corr(rsi_7)
corr_open_rsi_14 = open_price.corr(rsi_14)

# Create the scatter plot
plt.figure(figsize=(10, 6))

# Scatter plot for open price vs RSI 7
plt.scatter(open_price, rsi_7, color='blue', label=f'RSI (7-day), r = {corr_open_rsi_7:.2f}')

# Scatter plot for open price vs RSI 14
plt.scatter(open_price, rsi_14, color='green', label=f'RSI (14-day), r = {corr_open_rsi_14:.2f}')

# Titles and labels
plt.title('Open Price vs RSI (7-day and 14-day)')
plt.xlabel('Opening Price')
plt.ylabel('RSI')
plt.grid(True)

# Add a legend box on the side
plt.legend(title='Legend', bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)

# Adjust layout to make space for the legend box
plt.tight_layout()

# Show the plot
plt.show()
```
