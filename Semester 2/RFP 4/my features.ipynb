```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from sklearn.preprocessing import StandardScaler
```

```python
# Select numerical features for clustering
features = ["danceability", "energy", "tempo", "valence", "hype_score", "year"]

# Standardize the data
scaler = StandardScaler()
usable_data_scaled = scaler.fit_transform(usable_data[features])
```

```python
# Define range of cluster numbers to test
k_values = range(2, 11)  # Testing k from 2 to 10
silhouette_scores = []

# Loop through each value of k and calculate silhouette score
for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42)
    cluster_labels = kmeans.fit_predict(usable_data_scaled)
    
    # Calculate silhouette score
    score = silhouette_score(usable_data_scaled, cluster_labels)
    silhouette_scores.append(score)

# Plot the silhouette scores
plt.figure(figsize=(8, 5))
plt.plot(k_values, silhouette_scores, marker="o", linestyle="-", color="b")
plt.xlabel("Number of Clusters (k)")
plt.ylabel("Silhouette Score")
plt.title("Silhouette Score vs. Number of Clusters")
plt.grid(True)
plt.show()
```

```python
best_k = k_values[np.argmax(silhouette_scores)]  # Get the k with highest score

# Apply K-Means with optimal k
kmeans = KMeans(n_clusters=best_k, random_state=42)
usable_data["kmeans_cluster"] = kmeans.fit_predict(usable_data_scaled)

# Check cluster distribution
print(usable_data["kmeans_cluster"].value_counts())

# Display updated dataset
display(usable_data.head())
```
